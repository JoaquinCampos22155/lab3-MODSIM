# Escenario 1

Un país con una cobertura de vacunación contra la COVID-19 del 60 % observa un aumento del Rₜ de 0,8 a 1,5 tras la aparición de una nueva variante.

---

## 1. Factores que probablemente causaron este aumento

1. **Mayor transmisibilidad de la variante**  
   La nueva cepa porta mutaciones que facilitan la infección y el contagio entre personas (p. ej., mayor carga viral, transmisión más eficiente).

2. **Evasión inmunitaria o reducción de la eficacia vacunal**  
   La variante podría eludir parcial o totalmente la respuesta inmunitaria generada por las vacunas originales, reduciendo su protección efectiva y aumentando la proporción de susceptibles.

---

## 2. Propuestas de intervención para reducir el Rₜ

| **Tipo de intervención** | **Ejemplo concreto**                                                                                                                                                                                                                          |
|--------------------------|------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| **Económica**            | - **Incentivos financieros por refuerzo:** Bonos, vales o descuentos en servicios (transporte, alimentos) al completar la dosis de refuerzo.<br>- **Subsidio a teletrabajo y licencia remunerada:** Apoyo financiero a empresas para habilitar teletrabajo y garantizar pago durante aislamiento. |
| **No económica**         | - **Refuerzo de medidas NPI:** Obligatoriedad de mascarillas de alta filtración (FFP2/KF94) en interiores y transporte, limitación de aforos y mejora de ventilación.<br>- **Campaña de comunicación focalizada:** Mensajes claros sobre distanciamiento, lavado de manos y autoaislamiento ante síntomas. |


# Escenario 2

La tasa de vacunación contra el VPH es del 40 % en una región con una asistencia escolar del 80 %.  
¿Sería más eficaz un mandato de ingreso a la escuela o clínicas móviles gratuitas? Justifique con:

- **Umbral de inmunidad de grupo** (R₀ del VPH ≈ 4)  
- **Principios de la teoría de juegos**

---

## 1. Umbral de inmunidad de grupo

- R₀ ≈ 4  
- Umbral:  
  $$U = 1 - \frac{1}{R_0} = 1 - \frac{1}{4} = 0{,}75\;(75\%)$$  
- Cobertura actual: 40 %  
- Brecha para inmunidad:  
  $$75\% - 40\% = 35\;\text{puntos porcentuales}$$

---

## 2. Cobertura potencial de cada intervención

| **Intervención**                     | **Población cubierta / acceso** | **Cobertura esperada**             |
|--------------------------------------|---------------------------------|------------------------------------|
| **Mandato de ingreso a la escuela**  | 80 % de asistencia escolar      | Hasta 80 % (→ supera umbral)       |
| **Clínicas móviles gratuitas**       | Toda la población, voluntario   | Voluntad individual baja → < 75 %   |

---

## 3. Principios de la teoría de juegos

1. **Juego de coordinación**  
   - Con mandato escolar, el equilibrio es único: vacunación obligatoria a la entrada, elimina la incertidumbre sobre la estrategia colectiva.  
   - Con clínicas móviles, existen múltiples equilibrios y riesgo de **free-riding** (esperar que otros asuman el coste social).

2. **Externalidad negativa y free-riding**  
   - Sin mandato, cada individuo decide en función de su propio beneficio, subestimando el beneficio colectivo → **ineficiencia**.  
   - El mandato internaliza la externalidad: quien quiera asistir a la escuela debe vacunarse, garantizando el umbral.

---

## Conclusión

Un **mandato de ingreso a la escuela** es más eficaz porque:

- Garantiza una cobertura ≥ 75 % (umbral de inmunidad), dado el 80 % de asistencia.  
- Resuelve el problema de coordinación y free-riding descrito por la teoría de juegos.
