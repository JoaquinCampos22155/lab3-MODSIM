## Ejercicio 1

### 1. Retraso de primer orden vs. cadena de retrasos

- **Retraso de primer orden:** Retraso donde el efecto de un cambio no se ve de inmediato, sino que se ajusta gradualmente con el tiempo. 

- **Cadena de retrasos:** Es cuando el retardo se divide en varios pasos consecutivos. 

**Comparación:** El retraso de primer orden es un solo paso de ajuste lento, mientras que una cadena de retrasos simula un ajuste más suave al dividir ese cambio en varias partes.

---

### 2. Fuerza de infección (λ) vs. número básico de reproducción (R₀)

- **Fuerza de infección (λ):** Es la velocidad con la que una persona en riesgo se infecta. Depende de cuántos infectados hay en ese momento.

- **R₀ (número básico de reproducción):** Es la cantidad promedio de personas que infectará una persona enferma en una población susceptible.

**Comparación:** λ cambia con el tiempo según la cantidad de personas enfermas, mientras que R₀ es un valor fijo al inicio de una epidemia y se usa para predecir cuán contagiosa es una enfermedad.

---

### 3. ¿Por qué el sarampión necesita más del 95% de cobertura de vacunación, y el ébola solo el 50%?

Esto ocurre porque el **sarampión es mucho más contagioso** que el ébola. Una persona con sarampión puede infectar a más de 15 personas (R₀ muy alto), mientras que una persona con ébola contagia a 1 o 2. 

---

### 4. ¿Por qué las exigencias de vacunación alteran los equilibrios de Nash?

En teoría de juegos, las personas toman decisiones pensando en lo que harán los demás. Si la vacunación es obligatoria, las reglas del "juego" cambian: ya no se trata solo de decidir si vacunar o no, porque ahora hay consecuencias legales o sociales por no hacerlo. Esto cambia las estrategias y puede llevar a un comportamiento colectivo más cooperativo.

---

### 5. Lotería de vacunas de Ohio (conductual)

Este fue un incentivo donde las personas vacunadas participaban en una rifa con premios grandes. La idea era motivar a quienes dudaban vacunarse usando un premio como estímulo, aprovechando cómo nos atraen las recompensas rápidas.

---

### 6. Exigencias de ingreso escolar en Francia (estructural)

En Francia, los niños deben estar vacunados para entrar a la escuela. Esto no es un incentivo, sino una condición. Si no se cumple, no se puede acceder a ciertos servicios. Este enfoque cambia el entorno en el que se toma la decisión y empuja a las familias a vacunar por obligación.

**Comparación:** La lotería usa la emoción y motivación individual, mientras que las reglas escolares usan la estructura del sistema para garantizar cobertura.
